# Agrupando Dados

Aplica funções de sumarização a grupos de dados. 

```python
gb = df.groupby('DEPARTMENT').count()
gb.head()
```

```python
gb = df.groupby('DEPARTMENT')[['RACE', 'GENDER']].count()
gb.loc[('Library', )]
```

```python
gb = df.groupby(['DEPARTMENT', 'RACE'])['BASE_SALARY'].mean()
gb.head()
```

```python
gb = df.groupby(['DEPARTMENT', 'RACE'])['BASE_SALARY'].agg(lambda s: s.mean())
gb.head()
```
